# Linear Algebra, Recalls and Python

```python
import numpy as np
import matplotlib.pyplot as plt
fig_size = [0]*2
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = [0,0,0]
y = [0,0,0]
z = [0,0,0]
u = [1,0,0]
v = [0,1,0]
w = [0,0,1]
ax.quiver(x,y,z,u,v,w, length=1, pivot='tail', color='black')
ax.set_xlim([-0.5,1])
ax.set_ylim([-0.5,1])
ax.set_zlim([-0.5,1])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
xv, yv, zv = (0.3, 0.3, 0.7)
ax.quiver([0],[0],[0], [xv],[yv],[zv], lw=1, length=0.81, color='red', pivot='tail')
ax.plot([xv, xv],[yv, yv],[zv, 0],lw=1.5, linestyle=':', color='blue')
ax.plot([xv, xv],[yv, 0],[0,0],lw=1.5, linestyle=':', color='blue')
ax.plot([xv, 0],[yv, yv],[0,0],lw=1.5, linestyle=':', color='blue')
ax.plot([xv,0], [yv, 0], [zv, zv], lw=1.5, linestyle=':', color='blue')

plt.show()
```
